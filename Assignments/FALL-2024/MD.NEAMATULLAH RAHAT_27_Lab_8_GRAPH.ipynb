```python
class GraphMatrix:
  def __init__(self,vertices):
    self.vertices = vertices
    self.adj_matrix = []
    for i in range(vertices):
      row = [0]* vertices
      self.adj_matrix +=[row]

  def add_edge(self,x,y, weight = 1):
    self.adj_matrix[x][y] = weight
    self.adj_matrix[y][x] = weight

  def vertex_with_max_degree(self):
    max_degree = 0
    vertex = -1
    for i in range(self.vertices):
      degree = 0
      for j in self.adj_matrix[i]:
        if j!= 0:
          degree+=1
      if degree>max_degree:
        max_degree = degree
        vertex = i
    return vertex, max_degree

  def max_edge_sum(self):
    max_sum = 0
    vertex = -1
    for i in range(self.vertices):
      edge_sum = 0
      for j in range(self.vertices):
        edge_sum += self.adj_matrix[i][j]
      if edge_sum > max_sum:
        max_sum = edge_sum
        vertex = i
    return vertex, max_sum

  def max_outgoing_edge_sum(self):
    max_sum = 0
    vertex = -1
    for i in range(self.vertices):
      outgoing_sum = 0
      for j in self.adj_matrix[i]:
        outgoing_sum += j
      if outgoing_sum > max_sum:
        max_sum = outgoing_sum
        vertex = i
    return vertex, max_sum

  def convert_to_undirected(self):
    for i in range(self.vertices):
      for j in range(self.vertices):
        if self.adj_matrix[i][j] != 0:
          self.adj_matrix[j][i]= self.adj_matrix[i][j]
    for row in self.adj_matrix:
      print(row)

vertices = 7
edges = [
    (0, 1, 4), (0, 2, 3), (1, 3, 2), (1, 4, 6), (2, 4, 5),
    (3, 5, 8), (4, 5, 7), (4, 6, 9), (5, 6, 1)
]

graph_matrix = GraphMatrix(vertices)
for x,y, weight in edges:
  graph_matrix.add_edge(x, y, weight)

# Task 1
print("Vertex with max degree (adjacency matrix):", graph_matrix.vertex_with_max_degree())

# Task 2
print("Vertex with max edge sum (adjacency matrix):", graph_matrix.max_edge_sum())

# Task 3
print("Vertex with max outgoing edge sum (adjacency matrix):", graph_matrix.max_outgoing_edge_sum())

# Task 4
print("Adjacency Matrix after conversion:")
graph_matrix.convert_to_undirected()
```
