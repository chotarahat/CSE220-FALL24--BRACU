```python
#Complete this constructor
class vehicleNodes:
  def __init__(self, brand, vehicle_type, rent, passenger, next = None):
    self.brand = brand
    self.vehicle_type = vehicle_type
    self.rent = rent
    self.passenger = passenger
    self.next = next
```

```python
class VehicleHashTable:
  def __init__(self, size):
    self.vehicleTable = [None]*size
    self.empty_slot = {}
    for i in range(size):
      self.empty_slot[i] = False

  def create_from_vehicle_info_array(self, arr):
    for i in arr:
      self.insert_vehicle(i)

  def print_vehicle_hashtable(self):
    idx = 0
    for i in self.vehicleTable:
      print(idx, ':', end = ' ')
      head = i
      while head != None:
        print(f'(Brand: {head.brand}, Type: {head.vehicle_type}, Rent: {head.rent}, No. of Passengers: {head.passenger})', end = '---->')
        head = head.next
      print('None')
      print()
      idx += 1

  def find_empty_slot(self):
    for k,v in self.empty_slot.items():
      idx = k
      del self.empty_slot[k]
      return idx


  #Do it by yourself
  def __hash_function(self, brand):
    ascii_sum = sum(ord(char) for char in brand)
    index = ascii_sum % len(self.vehicleTable)



    # Do not discard the next 3 lines
    if index in self.empty_slot:
      del self.empty_slot[index]
    return index


  #Do it by yourself
  def insert_vehicle(self, vehicle):
    #TO DO
    brand,vehicle_type, rent, passenger = vehicle
    index = self.__hash_function(brand)
    new_node = vehicleNodes(brand, vehicle_type,rent,passenger)

    if self.vehicleTable[index] is None:
      self.vehicleTable[index] = new_node
    else:
      current = self.vehicleTable[index]
      if current.brand == brand:
        new_node.next = current
        self.vehicleTable[index] = new_node
      else:
        empty_idx = self.find_empty_slot()
        self.vehicleTable[empty_idx] = new_node

    return self.vehicleTable
```

```python
class Node_pair:
  def __init__(self, key, value, next = None):
    self.key, self.value, self.next = key, value, next


class Hashtable:
  def __init__(self, size):
    self.ht = [None]*size


  def insert(self, s):
    if self.search_hashtable(s) == 'Found':
      print(s,'Already Inserted. Cannot reinsert.')
      print('===============================')
      return
    hashed_index = self.__hash_function(s[0])
    pair = Node_pair(s[0], s[1])
    if self.ht[hashed_index] == None:
      self.ht[hashed_index] = pair
    else:
      pair.next = self.ht[hashed_index]
      self.ht[hashed_index] = pair


  def create_from_array(self, arr):
    for i in arr:
      self.insert(i)

  def print_hashtable(self):
    idx = 0
    for i in self.ht:
      print(idx, ':', end = ' ')
      head = i
      while head != None:
        print(f'(key: {head.key}, value: {head.value})', end = '-->')
        head = head.next
      print('None')
      print()
      idx += 1


  #Do it by yourself
  def __hash_function(self, key):
    #TO DO
    if len(key) %2 != 0:
      key += "N"
    ascii_sum = 0
    for i in range(0, len(key),2):
      pair_ascii = str(ord(key[i])) + str((ord(key[i+1])))
      ascii_sum += int(pair_ascii)
    return ascii_sum % len(self.ht)




  #Do it by yourself
  def search_hashtable(self, s):
    #TO DO
    key, value = s
    hashed_idx = self.__hash_function(key)
    current = self.ht[hashed_idx]
    while current != None:
      if current.key == key and current.value == value:
        return 'Found'
      current = current.next
    return 'Not Found'

```

Implement the __hash_function() and insert() methods

```python
class Node:
  def __init__(self, value=None, next = None):
    self.value = value
    self.next = next

class HashTable:
  def __init__(self, length):
    n = Node()
    self.ht = [n] * length
    self.length = length

  def show(self):
    count = 0
    for i in self.ht:
      temp = i
      print(count, end=" ")
      while temp!=None:
        print (temp.value, end="-->")
        temp = temp.next
      count += 1
      print()


  #Do it by yourself
  def __hash_function(self, key):
    #TO DO
    if len(key) % 2 == 0:
      total = 0
      for i in range(0,len(key),2):
        total += ord(key[i])
    else:
      total = 0
      for i in range(1,len(key),2):
        total += ord(key[i])

    return total % self.length

  #Do it by yourself
  def insert(self, key, value):
    #TO DO
    new_node = Node((key,value))
    idx = self.__hash_function(key)

    if self.ht[idx] == None:
      self.ht[idx] = new_node
    else:
      temp = self.ht[idx]
      prev = None
      while temp != None and temp.value!=None and temp.value[1]> value:
        prev = temp
        temp = temp.next
      if prev == None:
        new_node.next = self.ht[idx]
        self.ht[idx] = new_node
      else:
        new_node.next = temp
        prev.next = new_node
```

```python
class Node_pair:
  def __init__(self, key, value, next = None):
    self.key, self.value, self.next = key, value, next


class Hashtable:
  def __init__(self, size):
    self.ht = [None]*size


  def insert(self, s):
    hashed_index = self.__hash_function(s[0])
    pair = Node_pair(s[0], s[1])
    if self.ht[hashed_index] == None:
      self.ht[hashed_index] = pair
    else:
      pair.next = self.ht[hashed_index]
      self.ht[hashed_index] = pair


  def create_from_array(self, arr):
    for i in arr:
      self.insert(i)

  def print_hashtable(self):
    idx = 0
    for i in self.ht:
      print(idx, ':', end = ' ')
      head = i
      while head != None:
        print(f'({head.key}, {head.value})', end = '-->')
        head = head.next
      print('None')
      idx += 1


  #Do it by yourself
  def __hash_function(self, key):
    #TO DO
    return (key+3) % len(self.ht)



  #Do it by yourself
  def remove(self, key):
    #TO DO
    hashed_idx = self.__hash_function(key)
    head = self.ht[hashed_idx]
    prev = None

    while head != None:
      if head.key == key:
        if prev == None:
          self.ht[hashed_idx] = head.next
        else:
          prev.next = head.next
        return
      prev = head
      head = head.next

```
